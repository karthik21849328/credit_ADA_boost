```python
#importing all the required libraries 

import numpy as np
import pandas as pd 
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
import seaborn as sns 
```

```python
#loading the datset 

df=pd.read_csv("credit.csv")
df=pd.DataFrame(df)
```

```python
#by the above info we can notice that there are more number of non-numerical columns
# converting them into numerical type

for feature in df.columns:
    if df[feature].dtype=='object':
        df[feature]=pd.Categorical(df[feature]).codes
```

```python
#splitting the datset into training and testing 

train_set=df.head(700)
test_set=df.tail(300)

train_labels=train_set.pop('default')
test_labels=test_set.pop('default')
```

```python
dt_model=DecisionTreeClassifier(criterion='entropy')
```

```python
dt_model.fit(train_set,train_labels)
```

```python
dt_model.score(test_set,test_labels)
```

Decision tree alone giving accuracy of 66%

```python
from sklearn.ensemble import AdaBoostClassifier
adb=AdaBoostClassifier( n_estimators=50)

```

```python
adb=adb.fit(train_set,train_labels)
```

```python
y_predict=adb.predict(test_set)
```

```python
adb.score(test_set,test_labels)
```

we can observe that accuracy increased from 66% to 74% after using boosting technique
